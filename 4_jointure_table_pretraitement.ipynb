```python
import pandas as pd
import numpy as np
import os
import joblib
from pathlib import Path
from functools import reduce
```

# Importation et jointure des 3  bases de données 


```python
path = Path("DATA_PRETRAITEMENT")
```

```python
df_membre, df_reward, df_transaction = [pd.read_csv(path/i) for i in os.listdir(path) if (Path(i).suffix == ".csv") & (i != 'df_segmentation.csv')] 
```

```python
df_to_be_merged = [df_membre, df_transaction, df_reward]
```

```python
#COPIE_MEMBERS_ID
df_segmentation_copy_sociodem = df_membre[['MEMBER_ID','AGE', 'GENDER', 'PROV']]
#df_segmentation_copy_sociodem
```

```python
# On remplace les NaN par 0 car cela signie que le membre n'a pas fait de réclamation
df_segmentation = reduce(lambda left, right : pd.merge(left, right, on='MEMBER_ID', how="left"), df_to_be_merged).fillna(0)
```

```python
df_segmentation
```

```python
##probleme les nan des colonnes socio ont ete modifies par zero effacer colonnes sociodemographiques
df_segmentation.drop(['AGE', 'GENDER', 'PROV'], axis=1, inplace=True)
df_segmentation
```

```python
#ajouter les variables socio a nouveau 
df_segmentation = pd.merge(df_segmentation, df_segmentation_copy_sociodem, on = "MEMBER_ID", how = "inner")
df_segmentation
```

```python
df_segmentation.shape
```

# Renommer variables

```python
df_segmentation.rename(columns={'RECENCY':'NB_J_DEPUIS_DERNIERE_TRANSACTION', 'POINTS_REDEEMED_PCT_CHANGE_2021':'PTS_RDMED_PCT_CHANGE_2020_2021','NUMBER_CATEGORIES_PURCHASED_FROM': 'NB_CATG_PURCHASED_FROM'}, inplace=True)
```

```python
df_segmentation.columns = [i.upper() for i in df_segmentation.columns]
```
